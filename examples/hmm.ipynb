```python
import json

from pyeed.core import ProteinInfo, Alignment
from pyeed.align import ClustalOmega, HMM
```

```python
# Load sequences
with open("ids.json", "r") as f:
    ids = json.load(f)

sequences = ProteinInfo.get_ids(ids)
```

```python
# get last 30 sequences
alignment_sequences = sequences[-30:]

# make alignment with last 30 sequences
alignment = Alignment.from_sequences(alignment_sequences, ClustalOmega)
```

```python
# basic alignment visualization
alignment.show()
```

## Build HMM and serarch with HMM

```python
hmm = HMM(alignment=alignment)
hits = hmm.search(sequences)
```
